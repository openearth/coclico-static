```python
import xarray as xr
import numpy as np
import pandas as pd
import zarr
```

```python
np.random.seed(42)
waterlevel = np.random.randn(2, 2, 3)
lon = [[2.1, 2.2], [2.1, 2.2]]
lat = [[53, 53], [53.1, 53.1]]
t = pd.date_range("2022-01-01", periods=3)
reference_time = pd.Timestamp("2022-01-01")
```

```python
ds = xr.Dataset(
    data_vars=dict(
        waterlevel=(["x", "y", "time"], waterlevel),
    ),
    coords=dict(
        lon=(["x", "y"], lon),
        lat=(["x", "y"], lat),
        time=t,
        reference_time=reference_time,
    ),
    attrs=dict(description="Waterlevels."),
)
ds
```

```python
ds.to_zarr('./waterlevel')
```
